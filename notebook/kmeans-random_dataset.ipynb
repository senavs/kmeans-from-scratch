### Loading dataset from sklearn

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
dataset = make_blobs(n_samples=1000, centers=5, random_state=1)
```

```python
X = dataset[0][:,0]
Y = dataset[0][:,1],
LABELS = dataset[1]
```

### Target clusters

```python
plt.scatter(X, Y, c=LABELS)
plt.show()
```

### K Nearest Neighbors from scratch

```python
import os
import sys

# fixing import on jupyter
knn_module = os.path.abspath(os.path.join('..'))

if knn_module not in sys.path:
    sys.path.append(knn_module)
```

```python
from model.kmeans import KMeans

# fit
kmeans = KMeans(k=5, seed=101)
kmeans.fit(dataset[0], epochs=1)

# predict
predict_labels = kmeans.predict(dataset[0])
```

### Data validation

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10, 6))

ax1.set_title('original')
ax1.scatter(X, Y, c=LABELS)

ax2.set_title('predicted')
ax2.scatter(X, Y, c=predict_labels)
ax2.scatter(kmeans.centroids[:,0], kmeans.centroids[:,1], c='red', label='Centroids')

fig.legend()
plt.show()
```
